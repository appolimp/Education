```python
import numpy as np
from matplotlib import pyplot as plt, cm
import scipy
```

```python
def f(x):
    return np.sin(x / 5) * np.exp(x / 10) + 5 * np.exp(-x / 2)
```

```python
x = np.arange(1, 15.5, 0.5)
y = f(x)

plt.plot(x, y, '--')
plt.show()
```

```python
A_1 = np.array([[1**0, 1**1],
               [15**0, 15**1]])

B_1 = np.array([f(1), f(15)])

x_1 = np.linalg.solve(A_1, B_1)
print('a', A_1)
print('b', B_1)
print('x', x_1)

def f1(x):
    return x_1[0] + x_1[1] * x

plt.plot(x, y, '--', x, f1(x))
plt.show()
```

```python
A_2 = np.array([[1**0, 1**1, 1**2],
                [8**0, 8**1, 8**2],
                [15**0, 15**1, 15**2]])

B_2 = np.array([f(1), f(8), f(15)])

x_2 = np.linalg.solve(A_2, B_2)
print('a', A_2)
print('b', B_2)
print('x', x_2)

def f2(x):
    return x_2[0] + x_2[1] * x + x_2[2] * x**2

plt.plot(x, y, '--', x, f2(x))
plt.show()
```

```python
A_3 = np.array([[1**0, 1**1, 1**2, 1**3],
                [4**0, 4**1, 4**2, 4**3],
                [10**0, 10**1, 10**2, 10**3],
                [15**0, 15**1, 15**2, 15**3]])

B_3 = np.array([f(1), f(4), f(10), f(15)])

x_3 = np.linalg.solve(A_3, B_3)
print('a', A_3)
print('b', B_3)
print('x', x_3)

def f3(x):
    return x_3[0] + x_3[1] * x + x_3[2] * x**2 + x_3[3] * x**3

plt.plot(x, y, '--', x, f3(x))
plt.show()
```
