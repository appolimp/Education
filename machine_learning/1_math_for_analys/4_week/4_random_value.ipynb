```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
%matplotlib inline
```

```python
mu = 2.0
sigma = .5

norm_rv = sts.norm(loc=mu, scale=sigma)
norm_rv.rvs(size=10)
```

```python
norm_rv.cdf(3)
```

```python
x = np.linspace(0, 4, 100)
cdf = norm_rv.cdf(x)

plt.plot(x, cdf)
plt.show()
```

```python
norm_rv.pdf(2)
```

```python
x = np.linspace(0, 4, 100)
pdf = norm_rv.pdf(x)

plt.plot(x, pdf)
plt.show()
```

```python
a, b = 1, 4

uniform_rv = sts.uniform(a, b-a)
uniform_rv.rvs(10)
```

```python
x = np.linspace(0, 5, 100)
cdf = uniform_rv.pdf(x)

plt.plot(x, cdf)
plt.show()
```

```python
rv = sts.bernoulli(.7)

print(rv.rvs(10))

x = np.linspace(-1, 1, 5)
cdf = rv.cdf(x)
pdf = rv.pmf(x)

plt.plot(x, pdf, 'o', x, cdf)
plt.show()
```

```python
x = np.linspace(0, 30, 31)
for l in [1, 5, 10, 15]:
    rv = sts.poisson(l)
    cdf = rv.cdf(x)
    plt.step(x, cdf, label=f'$\lambda={l}$')

plt.legend()
plt.show()
```
