```python
from scipy import optimize
from matplotlib import pyplot as plt, cm
import numpy as np
```

```python
def rosen(x):
    return .5*(1-x[0])**2 + (x[1] - x[0]**2)**2

print(rosen([1, 1]))
```

```python
fig = plt.figure(figsize=[15, 10])
ax = fig.gca(projection='3d')

ax.view_init(45, 30)


X = np.arange(0.8, 1.2, 0.01)
Y = np.arange(0.8, 1.2, 0.01)

X, Y = np.meshgrid(X, Y)
Z = rosen(np.array([X, Y]))


surf = ax.plot_surface(X, Y, Z, cmap=cm.coolwarm)
plt.show()
```

```python
result = optimize.brute(rosen, ((-5, 5), (-5, 5)))
print(result)
```

```python
print(optimize.differential_evolution(rosen, ((-5, 5), (-5, 5))))
```

```python
def g(x):
    return np.array((-2*.5*(1 - x[0]) - 4*x[0]*(x[1] - x[0]**2), 2*(x[1] - x[0]**2)))
```

```python
print(optimize.check_grad(rosen, g, (2, 2)))
```

```python
print(optimize.fmin_bfgs(rosen, [2, 2], fprime=g))
```

```python
print(optimize.minimize(rosen, [2, 2], method='Nelder-Mead'))
```
